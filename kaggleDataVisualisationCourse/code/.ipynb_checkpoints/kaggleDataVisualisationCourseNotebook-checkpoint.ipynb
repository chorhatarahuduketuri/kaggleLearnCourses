## Interval data
Data that has meaningful order, like hurricane wind speed, shear strength of concrete, and the temperature of the sun, has _meaningful_ ordering, in that we can quantify what the difference between two entries is itself an interval variable. 

e.g. the difference between -20C water and +120C water is 140C worth of heat, x joules of energy, etc. i.e. an interval variable lies between them.

## Why subplot? 
#### Why are subplots useful? 
Often, when performing Exploratory Data Analysis (EDA) we make several graphs, perhaps of different types, exploring one or a small number of closely related aspects of the data. In these cases it makes a lot of sense to use subplotting to be able to see the relevant graphs side-by-side. 

Additionally, it is another element of style that can make visual presentation of the work we do more attractive to - and therefore easier to consume for - the consumer. 

Subplotting enables faceting. Faceting is when you break up data variables into different plots (perhaps so they can be seen more easily). When you still want them to be viewed together, you can use subplotting to present them side-by-side. `Seaborn` has a facility called `FacetGrid` that is useful for this, and will be looked at later in these tutorials.

## Countplot
The `pandas` bar chart becomes a `seaborn` `countplot`. 

```python
# Countplot of points
sns.countplot(reviews['points'])
```

## KDE Plot
`seaborn` doesn't have a direct analogue to the line or area chart, instead it has this. 

```python
# KDE plot of prices that are less than 200
sns.kdeplot(reviews.query('price < 200').price)
```

KDE is short for Kernel Density Estimate, which is a statistical technique for smoothing (hiding noise in) the data. 

This means graphs won't show irrelevant data you don't want it to, like if there was one wine priced $19.93 and a few hundred priced $20.00, the graph would have a pointless dip in it. 

See below for an example of what it's hiding. 

```python
# An unsmoothed line chart of prices < 200
reviews[reviews['price'] < 200]['price'].value_counts().sort_index().plot.line()
```

A KDE plot is better than a basic line chart for more usefully representing the data and what it means to a human. 

It's not always the best choice. It is worse for ordinal categorical data. For example, where there are wines that scored 85 points and other wines that scored 86 points, it would smooth the line to cross 85.5 between them, which is wrong because there is no 85.5 score. Use line charts in these cases instead. 

KDE plots can also come in 2 dimensions. 

```python
# 2 dimensional KDE plot of price and points features
sns.kdeplot(reviews[reviews['price'] < 200].loc[:, ['price', 'points']].dropna().sample(5000))
```

Bivariate KDE plots are very useful like scatter plots and hex plots, and are perhaps more visually appealing. Unfortunatley they are very computationally expensive, hence the limit of 5000 data samples for this graph. 

## Distplot
The `seaborn` equivalent to a `pandas` histogram is the `distplot`. 

```python
# Distplot of points
sns.distplot(reviews['points'], bins=10, kde=False)
```

## Scatterplot and hexplot
In `seaborn`, `jointplot` can be used to plot two variables against each other. 

```python
# Points and price jointplotted against each other
sns.jointplot(x='price', y='points', data=reviews[reviews['price'] < 100])
```

Other than the correlation coefficient provided (which can't be seen on this one for some reason), and the histograms, this is the same as the the basic `pandas` scatterplot. 

Passing `kind='hex'` makes it a hex plot. 

```python
# Hex plot of price vs points
sns.jointplot(
    x='price', 
    y='points', 
    data=reviews[reviews['price'] < 100], 
    kind='hex', 
    gridsize=20)
```

## Boxplot and violin plot
`seaborn` provides a boxplot fuction. It creates a statistically useful plot that looks like this. 

```python
# Boxplot of variety and points
df = reviews[
    reviews.variety.isin(reviews.variety.value_counts().head(5).index)
]

sns.boxplot(x='variety', y='points', data=df)
```

```python
# A violin plot of variety and points
sns.violinplot(
    x='variety',
    y='points',
    data=reviews[reviews.variety.isin(reviews.variety.value_counts()[:5].index)]
)
```

## Why seaborn?
`seaborn` is designed to work with tabled data out of the box. `pandas` is not - it requires your data to be made into an acceptable shape (`Series` and so on). 

Basically `seaborn` is better at working with data in the exact form it's likely to come out of a CSV file or a database, which is to say as a `DataFrame`.

```python
# Get data 
import pandas as pd
pd.set_option('max_columns', None)
df = pd.read_csv("../input/fifa-18-demo-player-dataset/CompleteDataset.csv", index_col=0)

import re
import numpy as np

footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])
```

```python
# Look at data
footballers.head()
```

```python
# Import seaborn
import seaborn as sns
```

## The FacetGrid
`FacetGrid` is an object which stores info on how to break up a data visualisation. e.g. comparing strikers and goalkeepers, we use `FacetGrid` to break `Position` down by `col` (column). 

```python
# Get data and make FacetGrid to fill in 
df = footballers[footballers['Position'].isin(['ST', 'GK'])]
g = sns.FacetGrid(df, col="Position")
```

```python
# Use map object method to plot data into grid
df = footballers[footballers['Position'].isin(['ST', 'GK'])]
g = sns.FacetGrid(df, col="Position")
g.map(sns.kdeplot, "Overall")
```

Notice the method (`sns.kdeplot`) being passed into the map call. 

If we want to put a lot of different features in, we can use `col_wrap` to help make it neat. 

```python
df = footballers

g = sns.FacetGrid(df, col="Position", col_wrap=6)
g.map(sns.kdeplot, "Overall")
```

Arranging things in grids: 

```python
df = footballers[footballers['Position'].isin(['ST', 'GK'])]
df = df[df['Club'].isin(['Real Madrid CF', 'FC Barcelona', 'Atlético Madrid'])]

g = sns.FacetGrid(df, row="Position", col="Club")
g.map(sns.violinplot, "Overall")
```

```python
# Same with custom ordering
df = footballers[footballers['Position'].isin(['ST', 'GK'])]
df = df[df['Club'].isin(['Real Madrid CF', 'FC Barcelona', 'Atlético Madrid'])]

g = sns.FacetGrid(df, row="Position", col="Club", 
                  row_order=['GK', 'ST'],
                  col_order=['Atlético Madrid', 'FC Barcelona', 'Real Madrid CF'])
g.map(sns.violinplot, "Overall")
```

### Why facet? 
Faceting is the easiest way to make data visualisations multivariate.

Pass it a `pd.DataFrame` in the right shape, and you get a gridded result of your variable values:

```python
# Prep stuff
import pandas as pd
pd.set_option('max_columns', None)
df = pd.read_csv("../input/fifa-18-demo-player-dataset/CompleteDataset.csv", index_col=0)

import re
import numpy as np

footballers = df.copy()
footballers['Unit'] = df['Value'].str[-1]
footballers['Value (M)'] = np.where(footballers['Unit'] == '0', 0, 
                                    footballers['Value'].str[1:-1].replace(r'[a-zA-Z]',''))
footballers['Value (M)'] = footballers['Value (M)'].astype(float)
footballers['Value (M)'] = np.where(footballers['Unit'] == 'M', 
                                    footballers['Value (M)'], 
                                    footballers['Value (M)']/1000)
footballers = footballers.assign(Value=footballers['Value (M)'],
                                 Position=footballers['Preferred Positions'].str.split().str[0])
```

```python
# Look at data
footballers.head()
```

## Adding more visual variables
Visual variables are any visual dimension or marker that we can use to perceptually distinguish two data elements from one another. E.g. size, colour, shape, 1/2/3 dimensional position. 

### Multivariate scatter plots

```python
import seaborn as sns

sns.lmplot(x='Value', y='Overall', hue='Position',
          data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])], 
          fit_reg=False)
```

```python
sns.lmplot(x='Value', y='Overall', markers=['o', 'x', '*'], hue='Position',
          data=footballers.loc[footballers['Position'].isin(['ST', 'RW', 'LW'])],
          fit_reg=False)
```

### Grouped box plot
**grouping**! 

```python
f = (footballers
     .loc[footballers['Position'].isin(['ST', 'GK'])]
     .loc[:, ['Value', 'Overall', 'Aggression', 'Position']]
    )
f = f[f['Overall'] >= 80]
f = f[f['Overall'] < 85]
f['Aggression'] = f['Aggression'].astype(float)

sns.boxplot(x='Overall', y='Aggression', hue='Position', data=f)
```

## Summarisation
Summarisation is the creation and addition of new variables by mixing and matching the information provided in the old ones. 

### Heatmap
The correlation plot. 

```python
f = (
footballers.loc[:, ['Acceleration', 'Aggression', 'Agility', 'Balance', 'Ball control']]
.applymap(lambda v: int(v) if str.isdecimal(v) else np.nan)
.dropna()
).corr()

sns.heatmap(f, annot=True)
```

### Parallel Coordinates

```python
from pandas.plotting import parallel_coordinates

f = (
footballers.iloc[:, 12:17]
.loc[footballers['Position'].isin(['ST', 'GK'])]
.applymap(lambda v: int(v) if str.isdecimal(v) else np.nan)
.dropna()
)
f['Position'] = footballers['Position']
f = f.sample(200)

parallel_coordinates(f, 'Position')
```
